```python
import pandas as pd
from bs4 import BeautifulSoup

# assuming your dataframe is named df

# Group the dataframe by 'rg' and sort by 'sequence' within each group
df = df.sort_values(by=['rg', 'sequence'])

# Initialize BeautifulSoup
soup = BeautifulSoup(features="html.parser")

# Iterate over the groups
for rg, group in df.groupby('rg'):
    # Initialize a new HTML document
    html = soup.new_tag('html')
    
    # Add a head section with some CSS
    head = soup.new_tag('head')
    style = soup.new_tag('style')
    style.string = """
    body {
        font-family: Arial, sans-serif;
        margin: 0 auto;
        width: 80%;
        padding: 20px;
    }
    h1 {
        color: #333;
    }
    p {
        line-height: 1.6;
    }
    """
    head.append(style)
    html.append(head)
    
    # Add a body section
    body = soup.new_tag('body')
    
    # Add a header with the rg number
    h1 = soup.new_tag('h1')
    h1.string = rg
    body.append(h1)
    
    # For each row in the group, create a new 'p' tag with the 'id' attribute set to the sequence number and the text set to the text field
    for _, row in group.iterrows():
        p = soup.new_tag('p', id=row['sequence'])
        p.string = row['text']
        body.append(p)
    
    # Append body to html
    html.append(body)
    
    # Construct the filename
    filename = f'../testimonies/{rg.replace(".", "_")}.html'
    
    # Write the HTML to a file
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(html.prettify())
```
